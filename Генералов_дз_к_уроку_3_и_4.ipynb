```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/svgen83/ml_python_blastim/main/weekly_clean.csv')
df
```

```python
df.loc[[1,5,10]]
```

```python
df[['Year','Region','New Cases']]
```

```python
min_cases = (
    df
    .groupby(['Region','Year'])
    .agg({'New Cases':'max'})
)
min_cases
```

```python
max_cases = (df
    .groupby(['Region','Year'])
    .agg({'New Cases':'min'})
)
max_cases
```

```python
df_mean = (
    df
    .groupby(['Region','Year'])
    .agg({'New Cases':np.mean})
)
```

```python
sum = (
    df
    .groupby(['Region','Year'])
    .agg({'New Cases':np.sum})
)
sum
```

```python
len(arabia[arabia['New Cases']==arabia_max])
```

```python
arabia_median = int(np.median(arabia_cases))
arabia_median
```

```python
len(arabia[arabia['New Cases']==arabia_median])
```
