```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('NTD_RAB2.csv')
names = df.iloc[0]
df.columns = names
df=df.drop(index=0)
df
```

```python
df = df.replace(['No data'], 0)
df
```

```python
ignore = ['Countries, territories and areas']

df = (df.set_index(ignore, append=True)
        .astype(float)
        .reset_index(ignore)
       )
df = df.set_index('Countries, territories and areas')
df
```

```python
df.describe()
```

```python
correlations = df.corr()
correlations
```

```python
plt.figure(figsize = (12,20))
df3 = df.iloc[:, 1:]
contries = df.iloc[:, 0:1]
matrix = np.triu(df3)

rp = sns.heatmap(
    data=df3,
    cbar= False,
    cmap= 'coolwarm',
    annot= True,
    mask=matrix    
).set_title('Rabies Death Cases')
rp
```

```python
df_t = df.T
df_t
```

```python
df_t.describe()
```

```python
sp = sns.scatterplot(df_t[['Tajikistan', 'Turkmenistan', 'Uzbekistan']])
sp.set_xlabel('Years')
sp.set_ylabel('Cases')
sp.set_title('Rabies in Middle_Asia_(post_USSR)')
sp.legend(bbox_to_anchor=( 1.02 , 1 ), loc='upper left')
```

```python
sp = sns.violinplot(df_t[['Tajikistan', 'Turkmenistan', 'Uzbekistan']])
sp.set_title('Rabies in Middle_Asia_(post_USSR)')
```
